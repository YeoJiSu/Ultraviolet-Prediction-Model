### 2020년 Raw Data에서 이상치 파악하기

```python
month = ["01","02","03","04","05","06","07","08","09","10","11","12"]
for i in range(0,12): # i는 0부터 11까지
  _month = month[i]
  df = pd.read_csv('/content/drive/MyDrive/Raw Data/2020/2020'+_month+'_uv.csv')
  uv=df['2020'+_month+'_uv.uv']
  date = df['2020'+_month+'_uv.yyyymmdd']
  new_date = set()
  count= 0
  stn_count = []
  for k in range(len(uv)):
    if uv[k] == -999:
      count+=1
      stn_count.append(df['2020'+_month+'_uv.stn'][k])
      new_date.add(date[k])
  print("-"*100)
  print("< "+str(i+1)+" 월 >")
  print("이상치 개수:", count)
  print("이상치가 나타나는 일 수:" ,len(new_date))
  print()
```

### 2020년도 이상치 제거하기

```python
month = ["01","02","03","04","05","06","07","08","09","10","11","12"]
for i in range(0,12): # i는 0부터 11까지
  _month = month[i]
  df = pd.read_csv('/content/drive/MyDrive/Raw Data/2020/2020'+_month+'_uv.csv')
  band1 = df['2020'+_month+'_uv.band1']
  print("-"*100)
  print("< "+str(i+1)+" 월 >")
  drop_list = []
  print("총 데이터 개수",len(df))

  # band1의 이상치 라인 삭제
  for j in range(0,len(df)):
    
    if band1[j] == -999 :
      drop_list.append(j)
  
  print("삭제된 개수", len(drop_list))
  new_df = df.drop(drop_list)
  new_df = new_df.drop('Unnamed: 0',axis=1)
  new_df = new_df.reset_index()

  # uv의 결측치 및 이상치 변경
  uv = new_df['2020'+_month+'_uv.uv']
  for k in range(0,len(new_df)):
    if k>0 and k < len(new_df)-1:
      if uv[k] >=0:
        continue
      else: 
        # uv: -999가 이고 위아래 정상적인 값이 존재할 때
        if uv[k-1]>=0 and uv[k+1]>=0:
          # 위아래 값의 평균값으로 이상치를 바꿔준다.
          mean = (uv[k-1]+uv[k+1])/2
          mean = round(mean,1)
          new_df.loc[k, '2020'+_month+'_uv.uv'] = mean

        # uv: 연속적인 -999가 일 때
        else:
          if uv[k-1]>=0: # uv [k-1]은 사실상 0보다 크거나 같을 수 밖에 없는데, 만약 앞에서 처리가 잘 안되었다면 if 문 통과가 안될 것임.
            index = k 
            while True:
              index+=1
              # + 가 나타나는 곳이 있으면 break
              if uv[index]>=0:
                break
            # 정상값 두개의 평균값으로 이상치를 바꿔준다.
            mean = (uv[k-1]+uv[index])/2
            mean = round(mean,1)
            new_df.loc[k, '2020'+_month+'_uv.uv'] = mean
          else:
            print(k) # k로 출력되는 게 없는 걸 보아하니 코드가 잘 동작하는 군
```
